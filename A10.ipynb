## Question 2 [8 marks]

Let $G$ be an undirected graph where every vertex has a loop. Show that the relation $R$ on the set of vertices of $G$ such that $vRu$ if and only if there is an edge associated to $\{v, u\}$ is a symmetric, reflexive relation on $G$.

YOUR ANSWER HERE

##### *written by Zitong Gu*

Let $u, v$ be arbitrary vertices of $G$. 

Assume that there is an edge associated to $\{u, v\}$. We want to show that $R$ is reflexive and symmetric. 

**Reflexive**: We must show that there exists a path from $u$ to itself. ($\forall u \in G [(u, u) \in R]$)

Since we were given that every vertex of $G$ has a loop, we know that there must exist a path from $u$ to itself. Thus, $R$ is reflexive. 

**Symmetric**: We must show that if there exists a path from $u$ to $v$ then there exists a path from $v$ to $u$. 

Since $G$ is an undirected graph which implies that there is an reversible path between $u$ and $v$, thus, we have shown that $R$ is symmetric.
